```python
#!pip install pandas
import numpy as np
import pandas as pd

exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
dataframe = pd.DataFrame(exam_data,index=labels)
dataframe
```

Task 4: Write a Pandas program to append a new row 'k' to data frame with these values(name : "Suresh", score: 15.5, attempts: 1, qualify: "yes")

```python
new_row = {"name" : "Suresh", "score": 15.5, "attempts": 1, "qualify": "yes"}
s=pd.Series(new_row).rename("k")
dataframe.append(s)
```

Task 6: Add a new column "Success" : if the score is higher than 10 we will have 1 else we will have 0

```python
def getList (d):
    res =[]
    for index, value in d.score.items(): 
        res.append( 1 if value> 10 else 0)
    return res
dataframe.assign(success=getList )
```
